```python
import numpy as np
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt

X_train = [[1], [2], [3], [4], [5], [6]] #2d vactors
y_train = [0, 0, 0, 1, 1, 1] #F,F,F,P,P,P. pass rate modified by me.

model = MLPClassifier(
    hidden_layer_sizes=(2,), #if to big will rely on present data too much not finding a regression/pattern
    solver='lbfgs', #adam big, sgd modify, lbfgs small
    max_iter=5000, #train times, if not enough will cause a warning
    random_state=42 #magic, stablizing. need to ask people what this mean.........
)

#train modelS
model.fit(X_train, y_train)

n=int(input("pick number"))

# Predict probability for n hours studied
prob = model.predict_proba([[n]])[0]

labels = ["Fail", "Success"]
values = [prob[0], prob[1]]
plt.bar(labels, values)
plt.ylabel("Probability")
plt.title(f"Pass vs Fail Probability ({n} hours study)")
plt.show()
```
